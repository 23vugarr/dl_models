```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('lithium-ion batteries.csv')
df['Has Bandstructure'] = np.where(df['Has Bandstructure'], 1, 0)
```

```python
target = np.array(pd.get_dummies(df['Crystal System']).astype('int64'))
features = df.drop(columns=['Materials Id', 'Formula', 'Spacegroup', 'Crystal System'])
features = np.array(features)

features = StandardScaler().fit_transform(features)

xtrain, xtest, ytrain, ytest = train_test_split(features, target, random_state=2, test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```python
model = Sequential()

model.add(Dense(1024, activation='relu', input_shape=(xtrain.shape[1],)))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(512, activation='relu', kernel_regularizer='l2'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(256, activation='relu', kernel_regularizer='l2'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])
```

```python
train = model.fit(
    xtrain, 
    ytrain,
    batch_size=32,
    verbose=1,
    epochs=100,
    validation_data=(xtest, ytest)
)
```
